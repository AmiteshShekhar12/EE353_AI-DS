`for each column, make a frequency plot to see if there is any major class imabalance`

```python
# Plot frequency of each column
for column in df.columns:
    if df[column].dtype == 'object':
        df[column].value_counts().plot(kind='bar', title=column)
        plt.show()
    else:
        df[column].plot(kind='hist', title=column, bins=30)
        plt.show()
```

`for the categorical data like job, loan and contact, we can check for class imbalance by analysing their frequency plot`\\
